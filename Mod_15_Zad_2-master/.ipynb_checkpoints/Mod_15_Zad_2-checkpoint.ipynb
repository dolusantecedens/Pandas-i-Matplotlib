```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import datetime as dt
df = pd.read_csv('HRDataset.csv')
df.dropna(thresh=2,inplace=True)
```

1. Nie istnieje zależność pomiędzy tym kto jest przełożonym pracownika a oceną jego pracy.

```python
plt.figure(figsize=(15,5))
sns.heatmap(df.pivot_table(index=df['PerformanceScore'],columns=df['ManagerName'],values='EmpID',aggfunc='count').fillna(0))
```

```python
plt.figure(figsize=(22,5))
sns.barplot(x='ManagerID',y='PerfScoreID',data=df)
```

2. Jeśli zależy nam na najdłuższym stażu pracy najlepszym źródłem pozyskania pracownika jest Careerbuilder.

```python
df['DOB'] = pd.to_datetime(df['DOB'])
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%Y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%Y')
```

```python
def count_seniority(row):
    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
df.pivot_table(index=df['RecruitmentSource'],values='Seniority',aggfunc='mean').sort_values(by="Seniority", ascending=False)
```

3. Zachodzi istotna lekka korelacja ujemna pomiędzy poziomem satysfakcji z pracy a byciem w związku małżeńskim.

```python
# df["MaritalDesc"]
plt.figure(figsize=(15,5))
sns.barplot(x='MaritalDesc',y='EmpSatisfaction',data=df)
```

```python
y = df['EmpSatisfaction']
x1 = df['MarriedID']
x = sm.add_constant(x1)
model = sm.OLS(y,x).fit()
model.summary()
```

4. Tak wygląda struktura wieku aktualnie zatrudnionych pracowników:

```python
df.columns
def count_age(row):
    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']
    return (end_date - row['DOB'])/np.timedelta64(1,'Y')
df["Age"] = df.apply(count_age ,axis=1)
df_prop = df[df["Age"] > 10]
df_actual = df_prop[pd.isnull(df["DateofTermination"])]
sns.displot(df_actual["Age"])
```

```python
sns.boxplot(y='Age',data=df_actual)
```

5. Starsi pracownicy nie pracują nad większą liczbą specjalnych projektów niż młodsi pracownicy.

```python
sns.jointplot(x='Age',y='SpecialProjectsCount',data=df_prop)
```

```python
sns.lmplot(x='Age',y='SpecialProjectsCount',data=df_prop,markers=['o'], scatter_kws={'s':50})
```

```python
y = df_prop['SpecialProjectsCount']
x1 = df_prop['Age']
x = sm.add_constant(x1)
model = sm.OLS(y,x).fit()
model.summary()
```
